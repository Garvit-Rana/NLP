```python
import pandas as pd
data=pd.read_csv("twitter_x_y_train.csv",delimiter=",")
data
y=data["airline_sentiment"]
x=data["text"]
from nltk.tokenize import word_tokenize
d=[]
for i in range(len(x)):
    d.append((word_tokenize(x[i])))
    
```

```python
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
l=WordNetLemmatizer()
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punct = list(string.punctuation)
stops.update(punct)
stops
```

```python
from nltk.corpus import wordnet
def get_simple(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean(w):
    out=[]
    for i in w:
        
        if i.lower() not in stops:
            p=pos_tag([i])
            g=l.lemmatize(i,get_simple(p[0][1]))
            out.append(g.lower())
    return out
```

```python
dc=[]
#i=0
for doc in d:
    a=(clean(doc))
    dc.append(a)
    print(len(dc))
```
