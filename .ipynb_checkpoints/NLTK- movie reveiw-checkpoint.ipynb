```python
from nltk.corpus import wordnet
def get_simple(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
l=WordNetLemmatizer()
from nltk.corpus import stopwords
import string
st=stopwords.words('english')
s=list(string.punctuation)
st=st+s

def clean(word):
    out=[]
    for i in word:
        if i.lower() not in st:
            p=pos_tag([i])
            clean=l.lemmatize(i,get_simple(p[0][1]))
            out.append(clean.lower())
    return out
```

```python
#doc=[(clean(doc),category) for doc,category in d]
dc=[]
#i=0
for doc,category in d:
    a=(clean(doc),category)
    dc.append(a)
    print(len(dc))
```

```python
train_data=dc[0:1500]
test_data=dc[1500:]
```

```python
doc=[]
for i in train_data:
    doc+=i[0]
```

```python
doc
```

```python
import nltk
f=nltk.FreqDist(doc)
```

```python
f
```

```python
com=f.most_common(3000)
```

```python
feature=[i[0] for i in com]
feature
```

```python
def get_feature(w):
    current={}
    word=set(w)
    for i in feature:
        current[i]=i in word
    return current
    
```

```python
train_d=[(get_feature(i),category) for i,category in train_data]
test_d=[(get_feature(i),category) for i,category in test_data]
```

```python
train_d[0]
```

```python
from nltk import NaiveBayesClassifier
```

```python
clf=NaiveBayesClassifier.train(train_d)
```

```python
nltk.classify.accuracy(clf,test_d)
```

```python
clf.show_most_informative_features(15)
```

```python
from sklearn.svm import SVC
from nltk.classify.scikitlearn import SklearnClassifier
```

```python
svm=SVC()
clf=SklearnClassifier(svm)
clf.train(train_d)
```

```python
nltk.classify.accuracy(clf,test_d)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
cls=SklearnClassifier(rf)
cls.train(train_d)
```

```python
nltk.classify.accuracy(cls,test_d)
```
